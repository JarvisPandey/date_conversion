```python
import re
dic={'jan ':'1','january':'1','feb ':'2','february':'2','mar ':'3','march':'3','apr ':'4','april':'4','may':'5','jun ':'6','june':'6','jul ':'7','july':'7','aug ':'8','august':'8','sep ':'9','sept ':'9','september':'9','oct ':'10','october':'10','nov ':'11','november':'11','dec ':'12','december':'12'}

dates={'one':'1','first':'1','second':'2','two':'2','three':'3','third':'3','four':'4','fourth':'4','five':'5','fifth':'5','six':'6','sixth':'6','seven':'7','seventh':'7','eight':'8','eighth':'8','nine':'9','nineth':'9','ten':'10','tenth':'10','eleven':'11','twelve':'12','twelfth':'12','thirteen':'13','thirteenth':'13','fourteenth':'14','fifteenth':'15','sixteenth':'16','fourteen':'14','fifteen':'15','sixteen':'16','seventeen':'17','seventeenth':'17','eighteen':'18','eighteenth':'18','nineteen':'19','nineteenth':'19','twenty':'20','twenty one':'21','twenty first':'21','twenty second':'22','twenty two':'22','twenty third':'23','twenty three':'23','twenty four':'24','twenty five':'25','twenty six':'26','twenty seven':'27','twenty eight':'28','twenty nine':'29','thirty':'30','thirty first':'31','thirty one':'31'}
```

```python
#s="13th dec "
t=0
for ii in df['sentence']:
    ii=ii+" "
    print(ii)
    prop=""
    date_int=0
    month_int=0

    date_num=re.search(r'(^([0-9]{1,2}))',ii)
    if date_num:
        a=date_num.group(1)
        #print(a)
        date_int=int(a)
        prop=prop+a+"-"
        
    date_text=re.search(r'(^((one|first|second|two|three|third|four|fourth|five|fifth|six|sixth|seven|seventh|eighth|eight|nine|nineth|ten|tenth|eleven|eleventh|twelve|twelfth|thirteen|thirteenth|fourteenth|fifteenth|sixteenth|seventeenth|fourteen|fifteen|sixteen|seventeen|eighteen|eighteenth|nineteen|nineteenth|twenty|twenty\sone|twenty\sfirst|twenty\ssecond|twenty\stwo|twenty\sthird|twenty\sthree|twenty\sfour|twenty\sfive|twenty\ssix|twenty\sseven|twenty\seight|twenty\snine|thirty|thirty\sfirst|thirty\sone)?(\s)?(-|\s)?(st|nd|rd|th|of)?(\s)?(jan\s|january|feb\s|february|mar\s|march|apr\s|april|may|jun\s|june|jul\s|july|aug\s|august|sep\s|sept|september|oct\s|october|nov\s|november|dec\s|december)))',ii)
    if date_text:
        date_out=date_text.group(3)
        for l,m in dates.items():
            if l==date_out:
                prop=prop+m+"-"
        
        
        
    month=re.search(r'(jan\s|january|feb\s|february|mar\s|march|apr\s|april|may|jun\s|june|jul\s|july|aug\s|august|sep\s|sept\s|september|oct\s|october|nov\s|november|dec\s|december)',ii)
    if month:
        b=month.group(1)
        #print("bbbbb",b)
        for i,j in dic.items():
            #print("iijjjjjjjjjjjj",i)
            if b==i:
                prop=prop+j+'-'
                month_int=int(j)
                #print(prop)
                break
    year=re.search(r'(([0-9]{1,4}(\s))$)',ii)
    if year:
        t=1
        yearr=year.group(0)
        prop=prop+yearr
        print(prop)
    if t!=1:
        monthh = datetime.datetime.now().strftime("%m")
        mon=int(monthh)
        yeear = datetime.datetime.now().strftime("%Y")
        yr=int(yeear)
        if mon>month_int:
            yr=yr+1
            yrr=str(yr)
            prop=prop+yrr
            print(prop)
        else:
            prop=prop+yeear
            print(prop)
```
